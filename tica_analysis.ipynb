```python
def eliminate_duplicated(keys, vals, n_dec=4):
    aux_vals = [round(v, n_dec) for v in freq.values()]
    new_keys = []
    new_vals = []
    for i, v in enumerate(aux_vals):
        try:
            if v == aux_vals[i+1]:
                new_keys.append(keys[i][:-2])
                new_vals.append(vals[i])

            else:
                if v != aux_vals[i-1]:
                    new_keys.append(keys[i])
                    new_vals.append(vals[i])

        except IndexError:
            pass

    return new_keys, new_vals
```

```python
sim = "5gjv_VSD_II_sin_S4_5_80A"
E = "0,145"
file = f"dataframes/data_{sim}_E={E}.csv"

exceptions = [1,27,94]
data = pd.read_csv(file)
data = data[~data["R"].isin(exceptions)]
data.set_index(['R', 'Time'], inplace=True)
data.drop(data.filter(like="msd"), axis=1, inplace=True)

freq = get_vars_freq(data, reps=100, n_contributions=5, criterion="pondered", lag=10000)
keys, values = eliminate_duplicated(list(freq.keys()), list(freq.values()))
keys = np.array(keys)
values = np.array(values)/max(values)

cut = 25

plt.figure(dpi=200, figsize=(7, 4))
# Crear el gráfico de barras
plt.bar(keys[:cut], values[:cut], color='blue', edgecolor='black', width=0.8)

# Añadir título y etiquetas a los ejes
#plt.title('Most Relevant Variables $E=0.16$', fontsize = 18)
plt.ylabel('Relevance Level', fontsize = 20)

# Rotar las etiquetas de los xticks
plt.xticks(rotation=45, ha = "right", size = 12)

plt.savefig(f"tica_params/tica2_{sim}_E={E}.png", bbox_inches='tight')

# Mostrar el gráfico
plt.show()
```

# AUTOCORRELATION ANALYSIS

```python
system = "5gjv_VSD_II_sin_S4_5_80A_E=0,145"

file = f"dataframes/data_{system}.csv"

data = pd.read_csv(file)
data.set_index(['R', 'Time'], inplace=True)
cols = ["rmsd_prot"]
df_imp = data[cols]
```

```python
lags = list(range(10000))
fig, ax = plt.subplots(dpi=200, figsize=(10,6))

for r in range(1, 30):
    autocorrelation_values = [df_imp[cols[0]].xs(r, level="R").autocorr(lag=lag) for lag in lags]
    ax.plot(lags, autocorrelation_values, label = "Rep %s" % r, linewidth=2)
ax.set_xlabel(r'$\tau_\textrm{lag}$', fontsize=20)
ax.set_ylabel(r'$R$', fontsize=25)
#ax.legend(fontsize = 18)
#ax.set_xlim([-20, 1000])
ax.set_ylim([-0.5, 1.05])
ax.grid()
plt.show()
#plt.savefig("../../Tesis/Images/langevin_based/correlation_k.png", bbox_inches='tight')
```

```python
cols = ["inertia_S2_om_3"]
df_imp = data[cols]

# Definir el valor umbral
threshold = 0.4
lags = list(range(10000))
tau_values = []


for r in range(1, 30):
    # Calcular la autocorrelación para cada lag
    autocorrelation_values = [df_imp[cols[0]].xs(r, level="R").autocorr(lag=lag) for lag in lags]

    # Encontrar el primer lag donde la autocorrelación cae por debajo del umbral
    tau = next((lag for lag, value in enumerate(autocorrelation_values) if value < threshold), None)
    tau_values.append(tau)
```
